```python
from matplotlib import pyplot
import numpy
%matplotlib inline
from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 16
```

```python
T=30.0
dt=0.03
nt=int(T/dt)
alpha=0.8
npart= 10000
sigma=1.0
mu=0.

x0 = numpy.random.normal(mu,sigma,npart)


binwidth=0.1
bins=numpy.arange(-5,5,binwidth)
bins = bins + binwidth/2
bc = bins[:-1]+binwidth/2


pyplot.xlim(-5,5)
pyplot.hist(x0,bins,normed=True);
pyplot.plot(bc, numpy.exp(-bc**2/(2.0*sigma**2))/numpy.sqrt(2*numpy.pi*sigma**2))
```

```python
T = 30.
dt =0.01
npart=10000
gamma = 0.5
nt= int(T/dt)
x = numpy.empty([nt+1,npart])
List=numpy.empty([nt+1,npart])
alpha_var=numpy.empty([2,9])
alpha_=numpy.empty([9])
x0=x0
for l in range (9) :
    alpha= 0.1 + l/10

    while int(nt/2) < i < nt:
        x[i+1]=x[i]+numpy.sqrt(2*alpha*dt)*numpy.random.normal(0,1,npart)-gamma*x[i]*dt
        List[i] = numpy.var(x[i])
    alpha_var[0,l]=numpy.average(List[1500:-5])
    alpha_var[1,l]=alpha
    binwidth = 0.5
bins = numpy.arange(-5.,5.,binwidth)
pyplot.plot(alpha_var[0],alpha_var[1]) 
numpy.polyfit(alpha_var[0],alpha_var[1],1)
```

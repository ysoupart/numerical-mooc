```python
from matplotlib import pyplot
import numpy
%matplotlib inline
from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 16

```

```python
a=numpy.random.sample(100000)
b=numpy.random.sample(100000)
z1= numpy.sqrt(-2*numpy.log(a))*numpy.cos(2*numpy.pi*b)
z2= numpy.sqrt(-2*numpy.log(a))*numpy.sin(2*numpy.pi*b)
    
binwidth=0.05
bins=numpy.arange(-3.,3., binwidth)
pyplot.title('Box-Muller Algorithm')
pyplot.xlim(-3,3)
pyplot.hist(z1,bins,normed=True, color='green');
pyplot.hist(z2,bins,normed=True, color='red');
xc = bins+(binwidth/2)
pyplot.plot(xc,numpy.exp(-xc**2/2.0)/numpy.sqrt(2*numpy.pi),lw=1.8,color='blue')

   
    
```
